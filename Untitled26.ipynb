```python
df['word_count']  = df['message'].apply(lambda x:utils._get_wordcounts(x) )
df['char_count']  = df['message'].apply(lambda x:utils._get_charcounts(x))
df['avg_wordlength']  = df['message'].apply(lambda x:utils._get_avg_wordlength(x))
df['stopwords']  = df['message'].apply(lambda x:utils._get_stopwords_counts(x))
df['dig_word_count']  = df['message'].apply(lambda x:utils._get_digit_counts(x))


```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df['message'],df['label'],test_size=0.2,random_state=42,stratify=df['label'])
```

```python
# Access the shape of individual elements within the tuple
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfd=TfidfVectorizer()
# Fit the vectorizer to the training data
x_train = tfd.fit_transform(x_train) # Fit and then transform the training data
# Transform the testing data using the fitted vocabulary
x_test = tfd.transform(x_test)
```

```python
x_train.toarray()
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
from ctypes import LibraryLoader
svc=LinearSVC()
svc.fit(x_train,y_train)
```

```python
y_predi=svc.predict(x_test)
```

```python
print(classification_report(y_test,y_predi))
```

```python
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from google.colab import drive

# Mount Google Drive (if necessary)
drive.mount('/content/drive')

# Load the model and TF-IDF vectorizer
with open('/content/drive/My Drive/spam_model.pkl', 'rb') as f:
    loaded_svc = pickle.load(f)

with open('/content/drive/My Drive/tfidf_vectorizer.pkl', 'rb') as f:
    loaded_tfd = pickle.load(f)

# Function to predict spam/ham
def predict_spam(text):
    # Transform the input text using the loaded TF-IDF vectorizer
    text_transformed = loaded_tfd.transform([text])

    # Make the prediction using the loaded model
    prediction = loaded_svc.predict(text_transformed)

    # Return the prediction
    if prediction[0] == 'ham':
        return "Not Spam"
    else:
        return "Spam"

# Get input from the user
user_input = input("Enter text to check for spam: ")

# Make the prediction and print the result
result = predict_spam(user_input)
print(result)
```
